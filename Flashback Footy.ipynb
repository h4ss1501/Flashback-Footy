Changing "tournament" column entries to highlight the year

```python
import pandas as pd

# Load the 'fifa_world_cup_matches.csv' file into a DataFrame
df = pd.read_csv('flashfooty_data/fifa_wc_matches.csv')

# Remove everything right of the decimal point in 'home_score' and 'away_score' columns
df['home_score'] = df['home_score'].astype(str).str.split('.').str[0]
df['away_score'] = df['away_score'].astype(str).str.split('.').str[0]

# Extract the year from the 'date' column
df['year'] = df['date'].str[:4]

# Ensure the 'year' column is of type string
df['year'] = df['year'].astype(str)

# Modify the 'tournament' column to have the value "FIFA {year}"
df['tournament'] = 'FIFA WC ' + df['year']

# Drop the temporary 'year' column
#df.drop(columns=['year'], inplace=True)

# Reordering the columns.
column_reorder = ['date', 'home_team', 'home_score', 
                  'away_score','away_team', 'tournament','country','city','year']
df = df[column_reorder]

# Display the modified DataFrame
df.head()

# Save the modified DataFrame to a new CSV file
df.to_csv('flashfooty_data/fifa_world_cup_matches_modified.csv', index=False)
```

Creating a new CSV file to filter out non world cup games

```python
import pandas as pd

# reading csv files
world_cup_df = pd.read_csv('flashfooty_data/fifa_world_cup_matches_modified.csv')
goalscorers_df = pd.read_csv('flashfooty_data/goalscorers.csv')

print("World cup DataFrame columns:")
print(world_cup_df.columns)
print("\nGoalscorers DataFrame columns:")
print(goalscorers_df.columns)
```

```python
# Loading both csv files

# World cup DataFrame
wc_df = pd.read_csv('flashfooty_data/fifa_world_cup_matches_modified.csv')

# Goalscorerers DataFrame
gs_df = pd.read_csv('flashfooty_data/goalscorers.csv')

# Merging the DataFrames on matches
merge_df = pd.merge(gs_df, wc_df, on=['date', 'home_team', 'away_team'], how='inner')

world_cup_goalscorers_df = merge_df[['date', 'home_team', 'away_team','team','scorer', 
                     'minute','tournament' ,'own_goal','penalty']]

world_cup_goalscorers_df.head()

world_cup_goalscorers_df.to_csv('flashfooty_data/fifa_world_cup_goalscorers.csv', index= False)
```

Penalty Shootouts for the world cup


```python
# World cup DataFrame
wc_df = pd.read_csv('flashfooty_data/fifa_world_cup_matches_modified.csv')

# Penalty shoout DataFrame
pen_shoot_df = pd.read_csv('flashfooty_data/shootouts.csv')

# Merging (inner merge) on DataFrames to select penalty shootouts on World Cup Games.
merge_df = pd.merge(pen_shoot_df, wc_df, on = ['date', 'home_team','away_team'], how= 'inner')

world_cup_penalties_df = merge_df[['date','home_team','away_team','winner','tournament','first_shooter']]

world_cup_penalties_df.to_csv('flashfooty_data/world_cup_penalties.csv', index= False) 
world_cup_penalties_df
```
